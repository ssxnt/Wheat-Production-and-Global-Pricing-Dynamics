```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('DCOILWTICO.csv')
```

```python
df['DATE'] = pd.to_datetime(df['DATE'])
```

```python
df = df.loc[df['DCOILWTICO'] != '.']
```

```python
df['DCOILWTICO'] = df['DCOILWTICO'].map(float)
```

```python
fig, ax1 = plt.subplots()
df.plot(x='DATE', y='DCOILWTICO', ax=ax1)
```
